#Slicing (Continuacion)

[inicio: stop:Incremento]

Nota:
1. El stop no es incluido
2. El incremento se puede modificar 
3. Soporta indexado alternativo (negativos)

```
frase="como hacer a la gente mas positiva?, muy facil, has que pierdan un electron!"
print(frase)
#Imprimir "ace" de la palabra hacer
print(frase [6: 9: 1])
#Valores por defecto

print(frase[ : 5: 2]) 

print(len(frase))

#Incremento
print(frase [ : :2]) #Toda la frase
print(frase[-1: : -1]) #Inverso -77

nombre="Juan Jose"
print(nombre[1: :-1])

primer=nombre[:4:]
print(primer)

print(nombre[-9: :1])
```

#Metodos de la clase string (str)

En python todo es un objeto (P.0.0.), eso significa que cada tipo de dato trae consigo metodos (funciones) incluidas.

Para identificar los metodos disponibles se puede usar una de dos funciones:
1. dir (str) #Resumen de los metodos de esa clase.
2. help (str) #Informacion completa de los metodos de esa clase


```
nombre="Juan Jose"
print(nombre)
print(dir (str))
```

Los nombre de metodos que incian con doble guion bajo significa que son metodos privados (P.O.O.), y no tendria porque usarlos como programador consumidor.

Los demas metodos son publicos y de libre uso.

```
print(nombre)
print(nombre.islower())
```

```
print(help(str))
```

```
nombre="PEDRO FERNANDEZ"
print(nombre.capitalize())#Minusculas
print(nombre.endswith("DEZ"))
print(nombre.replace("E","*"))#Sustituir una variable
ind_e=nombre.find("E")
print(ind_e)

print(nombre.replace("E","@",0))
print(nombre.lower())
print(nombre.upper())
nombre="jose jose el principe de la cancion"
print(nombre.count("jose"))
```

```
#Metodos para pricesar strings

numeros="2 , 1 , 3,4, 5 , 45          , 7"
print(numeros)

#Como extraigo los numeros de la cadena
#1. Eliminar los espacios al principio y al final con strip ()
numeros = numeros.strip()
print(numeros)


#2. Separo por comas.

numeros=numeros.split(",")
print(numeros)

#3. Con un for e indices limpiar cadenas de la lista
for indice in range(len(numeros) ):
    numeros[indice] = numeros[indice].strip()
print(numeros)

#4. Convertirlos a tipo entero
for indice in range(len(numeros)):
 numeros[indice] = int(numeros[indice]) 
print(numeros)

print(help(str.strip))
```
